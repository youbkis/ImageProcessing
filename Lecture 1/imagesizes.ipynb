```python
from PIL import Image
formats = ["png","jpg","tiff","gif","jp2"]
```

```python
#Imagefile to convert
imgname = 'images/schilder.jpg'
basename = imgname.split(".")[0]

im = Image.open(imgname)

outfiles = []
#Save new file formats
for f in formats:
    outfiles.append((basename + "." + f))
    im.save(basename + "." + f)

```

```python
#Get and Print filesizes
import os

print("original",f,round(os.path.getsize(imgname) / 1000),"KB","\n")

for f in outfiles:
    print(f,round(os.path.getsize(f) / 1000),"KB")
```

```python
#Show Images side by side
from matplotlib import pyplot
from matplotlib import image as pltimg
import matplotlib.cm as cm
import numpy as np
%matplotlib inline
fig= pyplot.figure(figsize=(150,150))

for i,f in enumerate(outfiles):
    pyplot.subplot(3, 4, i+1)
    pyplot.imshow(pltimg.imread(f),cmap=cm.gray)
    pyplot.axis('off')
    pyplot.title(f)
    pyplot.tight_layout()
pyplot.show()
```

```python
pyplot.clf()
pyplot.close()
```
